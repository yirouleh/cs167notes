## Exercise #3:
Write up a k-nearest-neighbors function like the one you made for the iris data set in class. It should be able to make mpg ("comb08") predictions for new cars based on the year, cylinders, and displacement. You should also be able to specify what you want to use as k.

Pay special attention to whether this problem is a classification or a regression, and how the answer to that question may change the output of your model. 

```python
#k-nearest-neighbors function
def knn(vechicle, data, k):
  data['distance_to_new'] = numpy.sqrt(
      (vechicle['year']-data.year)**2
      +(vechicle['cylinders']-data.cylinders)**2
      +(vechicle['displ']-data.displ)**2)
  
  sorted_data = data.sort_values(['distance_to_new'])
  return sorted_data.iloc[0:k]['comb08'].mean()
```

## Exercise #4: 
Demonstrate that your function works by making up some new values for hypothetical cars and using your function to display the predicted mpg ("comb08") for that car. 

```python
#random example of a vechicle 
vechicle = {}
vechicle['year'] = 1998
vechicle['cylinders'] = 8
vechicle['displ'] = 6.3

#print predicted mpg(comb08)
print("Predicted mpg: ", knn(vechicle, subset, 7))
```

## Exercise #5
Make a copy of the data and normalize the training data using Z-score-- `copy()` will be useful here. Predict the mpg of a particular car using your k-nearest-neighbors function with both the normalized and non-normalized training data. Compare your results. Use a markup cell to describe and explain the differences in a few sentences.

```python
# normalize a copy of the data
norm_subset = subset.copy()

#year
#calculate mean and std
y_mean = norm_subset['year'].mean()
y_std = norm_subset['year'].std()
norm_subset['year'] = (norm_subset['year'] - y_mean)/y_std

c_mean = norm_subset['cylinders'].mean()
c_std = norm_subset['cylinders'].std()
norm_subset['cylinders'] = (norm_subset['cylinders'] - c_mean)/c_std

d_mean = norm_subset['displ'].mean()
d_std = norm_subset['displ'].std()
norm_subset['displ'] = (norm_subset['displ'] - d_mean)/d_std

norm_subset.head()
```

```python
#non-normalized example
vechicle = {}
vechicle['year'] = 1998
vechicle['cylinders'] = 8
vechicle['displ'] = 6.3

#normalized example
z_vechicle = {}
z_vechicle['year'] = (vechicle['year'] - y_mean)/y_std
z_vechicle['cylinders'] = (vechicle['cylinders'] - c_mean)/c_std
z_vechicle['displ'] = (vechicle['displ'] - d_mean)/d_std

#5-NN
print("Predicted mpg with non-normalized data: ", knn(vechicle, subset, 5),
      "\nPredicted mpg with normalized data: ", knn(z_vechicle, norm_subset, 5))
```
